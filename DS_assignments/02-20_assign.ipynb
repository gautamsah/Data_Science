Q1. Explain GET and POST methods.

The GET Method
The GET method is used to request a resource (such as a web page) from the server. When you enter a URL in your web browser's address bar, it sends a GET request to the server, asking for the resource located at that URL. The data sent with a GET request is appended to the URL as a query string

- GET is used to request data from a specified resource.
- GET sends query string (name/value pairs) in the URL of a GET request with initial "?"
- GET requests can be cached
- GET requests remain in the browser history
- GET requests can be bookmarked
- GET requests should never be used when dealing with sensitive data
- GET requests have length restrictions
- GET requests are only used to request data (not modify)

The POST Method
The POST method is used to submit data to a server for processing. When you fill out a form on a web page and click the submit button, it sends a POST request to the server with the data you entered in the form. Unlike a GET request, the data sent with a POST request is not visible in the URL. Instead, it is sent in the request body.

- POST is used to send data to a server to create/update a resource.
- The data sent to the server with POST is stored in the request body of the HTTP request.
- POST requests are never cached
- POST requests do not remain in the browser history
- POST requests cannot be bookmarked
- POST requests have no restrictions on data length

Q2. Why is request used in Flask?

- The "request" object in Flask is used to represent the incoming HTTP request sent by a client to the server. 
- It provides access to various components of the request such as headers, cookies, query parameters, and form data.
- The "request" object is an instance of the "Request" class, which is part of the Flask framework. 
- When a client sends a request to a Flask application, the "Request" object is automatically created and populated with the relevant information from the request.
- Using the "request" object, you can extract information from the request and use it to generate a response. 
- request has following attributes: Form, args, Cookies, files, method

For example, you can retrieve data from a form submission, read query parameters, or inspect the headers of the incoming request. You can also use the "request" object to set cookies or modify the response headers.

Q3. Why is redirect() used in Flask?

In Flask, redirect() is used to redirect a user to a different route or URL with the specified status code. This function is particularly useful when you want to send the user to another page after a certain action has been performed.

An HTTP status code is a response from the server to the request of the browser. When we visit a website, a request is sent to the server, and the server then responds to the browser's request with a three-digit code: the HTTP status code. This status code also represents the error.

For example, suppose you have a login page on your website. After the user successfully logs in, you may want to redirect them to a page that displays their account information. To achieve this, you can use the redirect() function to send the user to the account information page.

```python
from flask import Flask, request

app = Flask(__name__)

@app.route('/greeting')
def greeting():
    name = request.args.get('name')
    if name:
        return f"Hello, {name}!"
    else:
        return "Please provide a name."

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
