Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web framework for Python. It is designed to make building web applications easier and more straightforward. Flask provides tools, libraries, and functionality for creating web applications using Python, which is a powerful programming language with a wide range of libraries and frameworks.

Some of the advantages of using Flask include:

1. Lightweight and flexible: Flask is a lightweight and flexible framework that allows developers to create web applications quickly and easily.

2. Minimalistic and easy to learn: Flask has a minimalistic approach to web development, making it easy to learn and understand. It doesn't require a lot of boilerplate code or configuration, so developers can get started quickly.

3. Modular design: Flask has a modular design that allows developers to add or remove features as needed. This makes it easy to customize the framework to suit your specific needs.

4. Extensive documentation and large community: Flask has extensive documentation and a large community of developers, making it easy to find answers to any questions or issues you may have.

5. Supports multiple database systems: Flask supports multiple database systems, including MySQL, PostgreSQL, SQLite, and others, making it easier to work with different data sources.

6. Built-in development server: Flask comes with a built-in development server that allows developers to test and debug their applications easily.

7. Scalable: Size is everything, and Flask’s status as a microframework means that you can use it to grow a tech project such as a web app incredibly quickly. If you want to make an app that starts small, but has the potential to grow quickly and in directions you haven’t completely worked out yet, then it’s an ideal choice. Its simplicity of use and few dependencies enable it to run smoothly even as it scales up and up.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

webapp=Flask(__name__)

@webapp.route("/home")
def helloWorld():
    return "<h1>Hello World!!!</h1>"

if __name__=="__main__":
    webapp.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of mapping URLs to specific functions or views in a Flask application. Flask uses the concept of app routes to handle incoming requests to specific URLs and to return the appropriate response.

When a user makes a request to a Flask application, the application needs to know which function or view to execute in order to generate the appropriate response. App routes are used to define which functions or views should be executed when certain URLs are requested. To define an app route in Flask, you can use the @app.route decorator.

e.g. 127.0.0.1:5000/hello

We use app routes because they allow you to organize your application's code and handle incoming requests in a structured and predictable way. By defining app routes, you can ensure that each URL in your application is handled by a specific function or view, making it easier to maintain and update your code over time.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask

webapp=Flask(__name__)

@webapp.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@webapp.route("/")
def details():
    return 'Company Name:ABC Corporation <br> Location: India <br> Contact Details: 999-999-9999'
if __name__=="__main__":
    webapp.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function is very useful for dynamically building a URL for a specific function. The function accepts the name of a function as first argument, and one or more keyword arguments, each corresponding to the variable part of URL.

```python
from flask import Flask, url_for

webapp=Flask(__name__)

@webapp.route('/')
def index():
    url = url_for('welcome')
    return f'The URL for welcome() function is: {url}'

@webapp.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

if __name__=="__main__":
    webapp.run(host="0.0.0.0")
```
